```python
"""
-- JOGO DA VELHA --
    Projeto desenvolvido para conclusão do curso Coding Tank da Let's Code.

    Natália Mafra | out/21
"""

def imprime_jogo():
    """Imprime o tabuleiro contendo as jogadas realizadas e os espaços disponíveis."""
    print(f"""
    {tabuleiro[0]} | {tabuleiro[1]} | {tabuleiro[2]}
    {tabuleiro[3]} | {tabuleiro[4]} | {tabuleiro[5]}
    {tabuleiro[6]} | {tabuleiro[7]} | {tabuleiro[8]}
    """)


def define_jogador():
    """Identifica quem é o jogador do turno e retorna as variáveis relacionadas a ele.""" 
    if contador_turnos % 2 == 0:
        jogador_turno = 1
        simbolo = 'O'
        jogadas_anteriores = posicoes_ocupadas_jogador1.copy()
    else:
        jogador_turno = 2
        simbolo = 'X'
        jogadas_anteriores = posicoes_ocupadas_jogador2.copy()
    
    return jogador_turno, simbolo, jogadas_anteriores


def faz_jogada():
    """Solicita ao jogador a posição da jogada e a retorna."""
    posicao_escolhida = input(f'Jogador {jogador_turno} | Insira o número da posição escolhida para jogar:')
    
    # Tabuleiro listado como string para a entrada não numérica 
    # não retornar erro, mas nova solicitação de input
    tabuleiro_string = list(map(str, tabuleiro))  

    while posicao_escolhida not in tabuleiro_string:
        posicao_escolhida = input('Digite um número disponível no espaço de jogo:')
    
    return int(posicao_escolhida)


def armazena_jogada():
        """Insere o símbolo do jogador no tabuleiro, armazena a posição e retorna a 
        lista de jogadas anteriores.
        """
        
        # Substitui número pelo símbolo no tabuleiro
        tabuleiro[tabuleiro.index(posicao_escolhida)] = simbolo  
        
        # Verifica o jogador do turno para escolher em que lista será adicionada
        # a posição escolhida nesta rodada
        if jogador_turno == 1:
            posicoes_ocupadas_jogador1.append(posicao_escolhida)
            jogadas_anteriores = posicoes_ocupadas_jogador1.copy()
        if jogador_turno == 2:
            posicoes_ocupadas_jogador2.append(posicao_escolhida)
            jogadas_anteriores = posicoes_ocupadas_jogador2.copy()
        
        return jogadas_anteriores

    
def condicao_de_vitoria():
    """Testa se as jogadas anteriores, inclusive a última, preenche uma das oito condições de vitória possíveis."""
    condicoes_vitoria = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [1, 4, 7], [2, 5, 8], [3, 6, 9], [1, 5, 9], [3, 5, 7]]
        
    for i in condicoes_vitoria:
        if (i[0] in jogadas_anteriores) and (i[1] in jogadas_anteriores) and (i[2] in jogadas_anteriores):
            return True
        
    return False


def jogar_novamente():
    """Permite ao usuário escolher jogar uma nova partida."""
    jogar_de_novo = input('\nDeseja jogar novamente? S/N ').upper()
    
    while (jogar_de_novo != 'S') and (jogar_de_novo != 'N'):
        jogar_de_novo = input('Digite uma opção válida (S/N) ').upper()
    
    if jogar_de_novo == 'S':
        return True
    
    print('Obrigada por jogar o jogo da velha!')
    return False
        
    
while True:
    
    # Início de jogo:
    tabuleiro = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    posicoes_ocupadas_jogador1 = []
    posicoes_ocupadas_jogador2 = []

    contador_turnos = 0

    imprime_jogo()
    
    # Até o 5º turno, não é necessário verificar se houve condição de vitória:
    while contador_turnos < 5:
        jogador_turno, simbolo, jogadas_anteriores = define_jogador()
        posicao_escolhida = faz_jogada()
        jogadas_anteriores = armazena_jogada()
        imprime_jogo()
        contador_turnos += 1
        
    # Passa a testar a condição de vitória e interrompe o programa se verificada:
    while contador_turnos < 9:
        if condicao_de_vitoria():
            break 
        jogador_turno, simbolo, jogadas_anteriores = define_jogador()
        posicao_escolhida = faz_jogada()
        jogadas_anteriores = armazena_jogada()
        imprime_jogo()
        contador_turnos += 1
    
    # Indica o resultado da partida:
    if condicao_de_vitoria():
        print(f'O Jogador {jogador_turno} ({simbolo}) venceu a partida!')
    else:
        print(f'Os jogadores empataram a partida!')
    
    # Questiona usuário sobre novo jogo:
    if not jogar_novamente():
        break        
```
