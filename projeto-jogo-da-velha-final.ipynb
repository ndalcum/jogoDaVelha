```python
# FUNÇÕES

def imprimeJogo():
    """
    Imprime o tabuleiro atual com as jogadas já realizadas e os espaços disponíveis numerados
    """
    print(f"""
    {tabuleiro[0]} | {tabuleiro[1]} | {tabuleiro[2]}
    {tabuleiro[3]} | {tabuleiro[4]} | {tabuleiro[5]}
    {tabuleiro[6]} | {tabuleiro[7]} | {tabuleiro[8]}
    """)
    
def defineJogador():
    """
    Define o jogador do turno e as variáveis relacionadas a ele de acordo com o número de turnos já realizados
    """
    jogador_turno = 1
    simbolo = 'O'
    jogadas_anteriores = posicoes_ocupadas_jogador1.copy()
   
    if contador_turnos % 2 != 0:
        jogador_turno = 2
        simbolo = 'X'
        jogadas_anteriores = posicoes_ocupadas_jogador2.copy()
    
    return jogador_turno, simbolo, jogadas_anteriores

def fazJogada():
    """
    Pede ao jogador que insira o número inteiro correspondente à posição escolhida para jogar e 
    o armazena numa lista
    """
    posicao_escolhida = input(f'Jogador {jogador_turno} | Insira o número da posição escolhida para jogar:')
    
    tabuleiro_string = list(map(str, tabuleiro))

    while (posicao_escolhida not in tabuleiro_string) or (posicao_escolhida.isalpha()): # Jogador repete a entrada enquanto a escolha for inválida
        posicao_escolhida = input('Digite um número disponível no espaço de jogo:')
    
    return int(posicao_escolhida)

def armazenaJogada():
        """ 
        Insere o símbolo do jogador na posição escolhida do tabuleiro e armazena a posição na lista de jogadas anteriores
        """
        tabuleiro[tabuleiro.index(posicao_escolhida)] = simbolo # Substitui número pelo símbolo no tabuleiro
        if jogador_turno == 1:
            posicoes_ocupadas_jogador1.append(posicao_escolhida) # Adiciona à lista de jogadas do Jogador 1 o número escolhido
            jogadas_anteriores = posicoes_ocupadas_jogador1.copy()
        if jogador_turno == 2:
            posicoes_ocupadas_jogador2.append(posicao_escolhida) # Adiciona à lista de jogadas do Jogador 2 o número escolhido
            jogadas_anteriores = posicoes_ocupadas_jogador2.copy()
        
        return jogadas_anteriores
            
def testaVitoria():
    """
    Testa se a última jogada preencheu uma das oito condições de vitória possíveis
    """
    
    condicoes_vitoria = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [1, 4, 7], [2, 5, 8], [3, 6, 9], [1, 5, 9], [3, 5, 7]]
        
    for i in condicoes_vitoria:
        if (i[0] in jogadas_anteriores) and (i[1] in jogadas_anteriores) and (i[2] in jogadas_anteriores):
            return True
        
    return False

def jogarNovamente():
    """
    Permite ao usuário escolher jogar uma nova partida
    """
    jogar_de_novo = input('\nDeseja jogar novamente? S/N ').upper()
    
    while (jogar_de_novo != 'S') and (jogar_de_novo != 'N'):
        jogar_de_novo = input('Digite uma opção válida (S/N) ').upper()
    
    if jogar_de_novo == 'S':
        return True
    
    print('Obrigada por jogar o jogo da velha!')
    return False
        
# PROGRAMA DO JOGO
while True:

    tabuleiro = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    posicoes_ocupadas_jogador1 = []
    posicoes_ocupadas_jogador2 = []
    vencedor = ''

    contador_turnos = 0

    imprimeJogo()

    while contador_turnos < 5:
        jogador_turno, simbolo, jogadas_anteriores = defineJogador()
        posicao_escolhida = fazJogada()
        jogadas_anteriores = armazenaJogada()
        imprimeJogo()
        contador_turnos += 1

    while contador_turnos < 9:
        if testaVitoria() == True:
            break 
        jogador_turno, simbolo, jogadas_anteriores = defineJogador()
        posicao_escolhida = fazJogada()
        jogadas_anteriores = armazenaJogada()
        imprimeJogo()
        contador_turnos += 1

    if testaVitoria() == True:
        print(f'O Jogador {jogador_turno} ({simbolo}) venceu a partida!')
    else:
        print(f'Os jogadores empataram a partida!')
    
    if jogarNovamente() == False:
        break        
```
