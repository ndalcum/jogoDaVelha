# Projeto - Jogo da Velha em Python

Chegou o momento de testarmos tudo o que aprendemos até aqui! Vamos criar um jogo da velha em Python  

Provavelmente você já sabe como o jogo funciona, mas vamos resgatar um pouco a nossa memória:  

O jogo é jogado em um tabuleiro 3x3 por dois jogadores. O primeiro jogador é representado pelo símbolo **O**, enquanto o segundo jogador utiliza o símbolo **X**. As jogadas são alternadas, preenchendo uma posição com o seu respectivo símbolo, sempre começando pelo jogador com o símbolo **O**. Ganha o jogador que conseguir colocar 3 símbolos em linha no tabuleiro, seja na vertical, horizontal ou diagonal.

O seu objetivo nesse projeto é criar um jogo da velha em Python, onde os jogadores poderão informar suas posições pelo teclado e o tabuleiro será mostrado com um print após cada jogada. Você pode utilizar qualquer recurso que aprendemos até aqui, além do seu conhecimento prévio. O único ponto importante é que __não será permitido o uso de bibliotecas__. 

### Avaliação

No dia 05/10/2021, cada aluno terá 3min30s para apresentar a solução em uma sala com os professores. A apresentação deve mostrar:

1. Uma partida com vitória do jogador com o símbolo **O** na horizontal
2. Uma partida com vitória do jogador com o símbolo **O** na vertical
3. Uma partida com vitória do jogador com o símbolo **X** na diagonal
4. Um empate
5. O código completo, de maneira superficial

O código da solução deverá ser submetido pelo Class como um arquivo. Serão aceitos arquivos nas extensões .py e .ipynb

Serão avaliados 3 aspectos em relação à entrega:

#### 1. Funcionalidade (5 pontos)

Verificaremos se o programa funciona: se é possível jogar de maneira alternada, sem que dois jogadores escolham uma mesma posição, e se os resultados dos jogos estão sendo computados de maneira correta.

#### 2. Clareza do código (2,5 pontos)

Analisaremos a escrita do código e a sua interpretabilidade. Pontos como a ausência de redundâncias, nomes de variáveis que façam sentido e comentários no código serão considerados como positivos.

#### 3. Explicação da solução (2,5 pontos)

Durante a apresentação, analisaremos a capacidade do aluno de explicar o código feito. Poderão ser feitos questionamentos simples sobre o código, como a explicação de uma linha.

**Importante**: plágios implicarão em nota zero para os envolvidos!
